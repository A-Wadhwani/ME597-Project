# Cleaning University of California Berkeley Power Consumption Dataset

The data retrieved from [UCB's energy dashboard](https://engagementdashboard.com/ucb/ucb) regarding power consumption in UCB's Mechanical Engineering building is in the form of a `json` format. Here we'll be converting it into a `csv` file for later data analysis. 


```python
import numpy as np
import pandas as pd
import seaborn as sns
import json
import pprint
from datetime import datetime
```

```python
filename = 'Datasets/Princeton Power Data/buildings-2019-'
df_power = pd.DataFrame()
for i in range(1,13):
    month_power = pd.read_csv(filename + str(i))
    df_power = df_power.append(month_power)
    
df_power = df_power[df_power['1915Hall.kWh'] != 'None']
df_power.to_csv('Datasets/Princeton_PowerConsumption.csv', index=False)
df_power.describe()
```
