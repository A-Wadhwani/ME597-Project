```python
#importing the needed libaries
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
weather_df = pd.read_csv("Plant_1_Weather_Sensor_Data.csv", parse_dates=['DATE_TIME'])
weather_df.head(10)
```

```python
#Dropping the unwanted columns
weather_needed_values_df = weather_df.drop(['PLANT_ID','SOURCE_KEY' ], axis =1)
weather_needed_values_df.describe()
```

```python
#Splitting the date time values 
weather_needed_values_df['Day'] = [d.day for d in weather_needed_values_df['DATE_TIME']]
weather_needed_values_df['Month'] = [d.month for d in weather_needed_values_df['DATE_TIME']]
weather_needed_values_df['Year'] = [d.year for d in weather_needed_values_df['DATE_TIME']]
weather_needed_values_df['Time'] = [d.time() for d in weather_needed_values_df['DATE_TIME']]
weather_needed_values_df['Date'] = [d.date() for d in weather_needed_values_df['DATE_TIME']]
weather_needed_values_df
```

```python
a_df = pd.DataFrame(weather_needed_values_df.groupby(['Day','Month' ]).mean().reset_index())
a_df.head()
```

```python
a = weather_needed_values_df['Date'].unique()
b = pd.DataFrame(a)
```

```python
weather_needed_values_df = weather_needed_values_df[weather_needed_values_df.IRRADIATION != 0.0]

weather_needed_values_df.head(10)
```

```python
sorted_a_df = a_df.sort_values(by=['Month', 'Day']).reset_index()
print(sorted_a_df)
```

```python
final_df = pd.concat([ sorted_a_df, b], axis = 1)
```
